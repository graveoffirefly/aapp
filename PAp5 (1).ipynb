```python
import pandas as pd
df=pd.read_csv("Sample - Superstore.csv",encoding='latin1')
df.head()
```

```python
df.columns
```

```python
df['Profit']=df['Profit'].astype('int64')
df['Segment']=df['Segment'].astype('category')
df[['Profit','Segment']].info()
```

```python
df['UnitPrice']=df['Sales']/df['Quantity']
df[['UnitPrice','Sales','Quantity']].head()
```

```python
df['DiscountPrice']=df['Discount']*df['Sales']
df[['DiscountPrice','Discount','Sales']].head()
```

```python
df['CostPrice']=df['Sales']-df['Profit']
df[['CostPrice','Sales','Profit']].head()
```

```python
import numpy as np
df['Sales_log']=np.log(df['Sales'])
df[['Sales_log','Sales']].head()
```

```python
df['Sales_log2']=np.log2(df['Sales'])
df[['Sales_log2','Sales']].head()
```

```python
df['Sales_log10']=np.log10(df['Sales'])
df[['Sales_log10','Sales']].head()
```

```python
df['Sales_rep']=np.reciprocal(df['Sales'])
df[['Sales_rep','Sales']].head()
```

```python
df['Quantity_sq']=np.power(df['Quantity'],2)
df[['Quantity','Quantity_sq']].head()
```

```python
df['Sales_sqrt']=np.sqrt(df['Sales'])
df[['Sales_sqrt','Sales']].head()
```

```python
df['initi']=df['Order ID'].str[:2]
df['initi']
```
