```python
import pandas as pd
df = pd.read_csv("Downloads/Loan - Loan.csv")
df.head()
```

```python
df.info()
```

```python
df['Date']=pd.to_datetime(df['date_issued'])
df.info()
```

```python
df['Month']=df['Date'].dt.month
df['date_issued:year'] = df['Date'].dt.year
df['day']=df['Date'].dt.day
```

```python
df[['day','Month','date_issued:year']].head()
```

```python
df['date_issued:day_of_week']=df['Date'].dt.day_of_week
df['date_issued:day_of_year']=df['Date'].dt.day_of_year
```

```python
df[['date_issued','date_issued:year',
    'date_issued:day_of_week','date_issued:day_of_year']].head()
```

```python
def week_part(day):
    if day in [1,2,3,4,5,6,7]:
        return "week 1"
    elif day in [8,9,10,11,12,13,14]:
        return "week 2"
    elif day in [15,16,17,18,19,20,21]:
        return "week 3"
    elif day in [22,23,24,25,26,27,28]:
        return "week 4"
    elif day in [29,30,31]:
        return "week 5"
```

```python
df ['Week_No']= df['day'].apply(week_part)
df[['day','Week_No']].head()
```

```python
import numpy as np
df['date_issued:day_of_weekend'] = np.where(df['date_issued:day_of_week'].isin([5, 6]), 1, 0)
df[['date_issued', 'date_issued:day_of_week', 'date_issued:day_of_weekend']].head()
```
